```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Path
from matplotlib.path import Path
import matplotlib.patches as patches

def lissajous(a, b, phi, num_points=1000, plot=True):
    """
    Generiert eine Lissajous-Figur und berechnet deren Länge.
    
    Parameter:
    a (float): Frequenz der x-Komponente
    b (float): Frequenz der y-Komponente
    phi (float): Phasenverschiebung in Radiant
    num_points (int): Anzahl der Punkte für die Berechnung
    plot (bool): Wenn True, wird die Figur geplottet
    
    Returns:
    tuple: (Kurvenlänge, (x-Koordinaten, y-Koordinaten))
    """
    # Zeitparameter
    t = np.linspace(0, 2*np.pi, num_points)
    
    # Berechnung der Koordinaten
    x = np.sin(a * t)
    y = np.sin(b * t + phi)
    
    # Berechnung der Kurvenlänge
    dx = np.diff(x)
    dy = np.diff(y)
    curve_length = np.sum(np.sqrt(dx**2 + dy**2))
    
    if plot:
        # Erstelle einen neuen Plot
        fig, ax = plt.subplots(figsize=(8, 8))
        
        # Plotte die Lissajous-Figur
        ax.plot(x, y, 'r-', linewidth=1.5)
        
        # Füge Koordinatenachsen hinzu
        ax.axhline(y=0, color='k', linestyle='-', alpha=0.3)
        ax.axvline(x=0, color='k', linestyle='-', alpha=0.3)
        
        # Setze die Grenzen und Aspektverhältnis
        ax.set_xlim(-1.2, 1.2)
        ax.set_ylim(-1.2, 1.2)
        ax.set_aspect('equal')
        
        # Beschriftungen
        ax.set_title(f'Lissajous-Figur (a:{a}, b:{b}, φ:{phi/np.pi:.2f}π)')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        
        # Füge Textbox mit Informationen hinzu
        info_text = f'Frequenzverhältnis: {a}:{b}\nKurvenlänge: {curve_length:.2f}'
        plt.text(0.02, 0.98, info_text,
                transform=ax.transAxes,
                verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        plt.grid(True, alpha=0.3)
        plt.show()
    
    return curve_length, (x, y)

# Beispielaufrufe
if __name__ == "__main__":
    # Beispiel 1: Einfache 1:1 Lissajous-Figur
    length1, _ = lissajous(1, 1, 0)
    print(f"Länge der 1:1 Kurve: {length1:.2f}")
    
    # Beispiel 2: 2:3 Lissajous-Figur mit Phasenverschiebung
    length2, _ = lissajous(2, 6, np.pi/4)
    print(f"Länge der 2:3 Kurve mit φ=π/4: {length2:.2f}")
    
```
